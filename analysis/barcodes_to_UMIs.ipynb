# Purpose
for each cell there is a unique 10X barcode. Prior to 10X sequencing we assigned the following barcodes to these corresponding cell populations:
pBA571-07 ; CTAACCTTTGTCACTCTG	SW480 parental
pBA571-06 ; CACGCCATCCTTGTTATT	SW480 LVM1
pBA571-09 ; GGTTCGGAGTAGCGGGAT	SW480 LVM2

The goal is to assign these barcodes, hereafter called 'GBCs' to their corresponding 10X barcode, hereafter called '10X_BCs' which identify each cell. This way we can assign individual cells to their population of origin.

The library design strategy is roughly as follows:
Read 1 (P5) reads the [[10X_BC - UMI]] ; 10X_BC = 16bp, UMI = 10 bp
Read 2 (P7) reads the transcript. In this case we are amplifying a portion of the transcript that contains the GBC. [[transcript - GBC - transcript]]

```python
import os
import pandas as pd
import numpy as np
import math
import Bio
from Bio.Seq import Seq
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
%matplotlib inline
import rpy2 as rpy2
%load_ext rpy2.ipython
```

```python
########first select for reads that match the perturb seq backbone amplicon site
head_constant = 'TAGCAAACTGGGGCACAAGCTTAATTAAGAATT'

#######trim off constant head
cmd = 'cutadapt -e 0.15 -pair-filter=any --untrimmed-output r2_notbarcode.fastq --untrimmed-paired-output r1_notbarcode.fastq -g ^'+head_constant+' -o out_qc.2.fastq -p out_qc.1.fastq barcode_amplify_S5_L002_R2_001.fastq.gz barcode_amplify_S5_L002_R1_001.fastq.gz '
print(cmd)
os.system(cmd)
```

```python
#########then select for reads that match individual barcodes

gbc_1 = 'CTAACCTTTGTCACTCTG'
gbc_2 = 'CACGCCATCCTTGTTATT'
gbc_3 = 'GGTTCGGAGTAGCGGGAT'

GBC = []

GBC.append(gbc_1)
GBC.append(gbc_2)
GBC.append(gbc_3)

for barcode in GBC:
    rev_comp = str(Seq(barcode).reverse_complement())
    cmd = 'cutadapt -e 0.15 -pair-filter=any --discard-untrimmed -g ^'+rev_comp+' -o '+barcode+'_out.2.fastq -p '+barcode+'_out.1.fastq out_qc.2.fastq out_qc.1.fastq'
    print(cmd)
    os.system(cmd)

###results: 98.9% of perturb seq selected reads contained barcodes (15% error rate)
###total coverage ~15% of reads contain a usable barcode
###remaining files containing reads start with the sequence of the barcode and out.1 = read1, out.2 = read2
###do autopsy on unmatched reads.
```

```python
%%bash
##############look at distributions prior to collapsing for each barcode; implement cutoff of # reads / barcode

###########get only barcode sequences
for file in BConly*fastq;
do
    newfile="dist_"$file
    awk '{if(NR%4==2) {print $0}}' $file > $newfile
done

###########count uniq 
for file in dist_*;
do
    newfile="counts_"$file
    sort $file | uniq -c | sort -n | awk '{print $1}' > $newfile
done
```

```python
%%bash

###########get only barcode sequences

for file in Clpsd_*;
do
    newfile="seqs_"$file
    awk '{if(NR%4==2) {print $0}}' $file > $newfile
done
```

```python
###########filter 10X barcodes from cellranger

###########format
BC_file = pd.read_csv('barcodes.tsv',sep='\t',header=None)
BC_file['BC'] = BC_file.apply(lambda row: str(row[0])[0:16],axis=1)
BC = BC_file.drop([0],axis=1)
BC = list(BC['BC'])
```

```python
###########select only barcodes that match 10X barcodes
def intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))
##############read in files for matching
CACG = 'seqs_Clpsd_BConly_CACGCCATCCTTGTTATT_out.1.fastq'
CTAA = 'seqs_Clpsd_BConly_CTAACCTTTGTCACTCTG_out.1.fastq'
GGTT = 'seqs_Clpsd_BConly_GGTTCGGAGTAGCGGGAT_out.1.fastq'
#####
with open(CACG) as f:
    CACG_BC = f.read().splitlines()
with open(CTAA) as f:
    CTAA_BC = f.read().splitlines()
with open(GGTT) as f:
    GGTT_BC = f.read().splitlines()
##########filter out barcodes from 'tmp_* above', if necessary
exCACG = 'tmp_dist_BConly_CACGCCATCCTTGTTATT_out.1.fastq'
exCTAA = 'tmp_dist_BConly_CTAACCTTTGTCACTCTG_out.1.fastq'
exGGTT = 'tmp_dist_BConly_GGTTCGGAGTAGCGGGAT_out.1.fastq'
#####
with open(exCACG) as f:
    exCACG_BC = f.read().splitlines()
with open(exCTAA) as f:
    exCTAA_BC = f.read().splitlines()
with open(exGGTT) as f:
    exGGTT_BC = f.read().splitlines()
#####
CACG_BC = list(set(CACG_BC)-set(exCACG_BC))
CTAA_BC = list(set(CTAA_BC)-set(exCTAA_BC))
GGTT_BC = list(set(GGTT_BC)-set(exGGTT_BC))
###############filter overlapping
CACG_uniq = list(set(CACG_BC)-set(CTAA_BC)-set(GGTT_BC))
CTAA_uniq = list(set(CTAA_BC)-set(CACG_BC)-set(GGTT_BC))
GGTT_uniq = list(set(GGTT_BC)-set(CACG_BC)-set(CTAA_BC))
##############match with barcode file
total = 0
###
CACG_out = intersection(CACG_uniq,BC)
print('CACG')
print(len(CACG_uniq))
print(len(CACG_out))
total += len(CACG_out)
###
CTAA_out = intersection(CTAA_uniq,BC)
print('CTAA')
print(len(CTAA_uniq))
print(len(CTAA_out))
total += len(CTAA_out)
###
GGTT_out = intersection(GGTT_uniq,BC)
print('GGTT')
print(len(GGTT_uniq))
print(len(GGTT_out))
total += len(GGTT_out)
###
print('total:',total)
print('numcells:',len(BC))
print('%',float(total)/len(BC))
```

```python
######write out final filtered barcodes
###
CACG_outfile = open('CACG_outfile.txt', 'w')
for item in CACG_out:
    CACG_outfile.write("%s\n" % item)
###
CTAA_outfile = open('CTAA_outfile.txt', 'w')
for item in CTAA_out:
    CTAA_outfile.write("%s\n" % item)
###
GGTT_outfile = open('GGTT_outfile.txt', 'w')
for item in GGTT_out:
    GGTT_outfile.write("%s\n" % item)
###
```
