```python
import pandas as pd
import os 
import numpy as np
import scipy as sp
import scipy.stats as stats
import Bio
import rpy2
%load_ext rpy2.ipython
```

```python
%%R

###
library(SingleCellExperiment)
library(scater)
library(TSCAN)
options(stringsAsFactors = FALSE)
library(plyr)
library(tidyverse)
###readin
scdata <- readRDS("scater/ready_for_clustering.rds")
```

```python
%%R

####get PCA vectors for PC1 and PC2 from scexperiment
PCA <- reducedDim(scdata, "PCA")
####strip attribute carried over from scexperiment
attr(PCA, "percentVar") <- NULL
####format
PCA <- data.frame(PCA)
master <- data.frame(PCA)
####add in sample data
master$sample <- colData(scdata)$sample
####read into 3 unique dataframes
P <- subset(master, sample == 'Parent') 
LVM1 <- subset(master, sample == 'Lvm1')
LVM2 <- subset(master, sample == 'Lvm2')
####save
write.table(PCA,file="pkg_writing/PCA_coord.tsv",sep='\t',col.names=FALSE)
```

```python
%%R
#####calculate distances
P_dist = as.matrix(dist(P))
LVM1_dist = as.matrix(dist(LVM1))
LVM2_dist = as.matrix(dist(LVM2))
PCA_dist = as.matrix(dist(PCA))
```

```python
%%R
####define radius
radius = 5
####count number within each radius for each cell
P_counts <- apply(P_dist, 2, function(x) length(x[x<1]))
LVM1_counts <- apply(LVM1_dist, 2, function(x) length(x[x<1]))
LVM2_counts <- apply(LVM2_dist, 2, function(x) length(x[x<1]))
PCA_counts <- apply(PCA_dist, 2, function(x) length(x[x<1]))
```

```python
%%R

####add labels and join counts
P_counts <- data.frame(P_counts)
P_counts$sample <- "P"
colnames(P_counts) <- c("counts","sample")
####
LVM1_counts <- data.frame(LVM1_counts)
LVM1_counts$sample <- "LVM1"
colnames(LVM1_counts) <- c("counts","sample")
####
LVM2_counts <- data.frame(LVM2_counts)
LVM2_counts$sample <- "LVM2"
colnames(LVM2_counts) <- c("counts","sample")
####
PCA_counts <- data.frame(PCA_counts)
```

```python
%%R
####concatenate tables
print(dim(LVM1_counts))
all_counts <- rbind(P_counts,LVM1_counts,LVM2_counts)
print(dim(all_counts))
####format
all_counts$barcode <- row.names(all_counts)
####add in total counts
PCA_counts$barcode <- row.names(all_counts)
####join
master <- join(all_counts,PCA_counts, by = "barcode")
####get percentage of within group by distance
master$pct_ingroup <- master$counts/master$PCA_counts
####join with PCA, PC1/PC2 coordinate values from prior
PCA$barcode <- row.names(PCA)
master <- join(master,PCA, by = "barcode")
```

```python
%%R

test <- master
test$pct_ingroup <- sqrt(test$pct_ingroup)

ggplot(data=test, mapping = aes(x=PC2,y=PC1)) +
    geom_point(mapping = aes(color=sample , alpha=pct_ingroup))
```

```python
%%R

test <- master
test$pct_ingroup <- sqrt(test$pct_ingroup)

####subset
test <- subset(test, sample=="P")

ggplot(data=test, mapping = aes(x=PC2,y=PC1)) +
    geom_point(mapping = aes(color=sample , alpha=pct_ingroup))









```
