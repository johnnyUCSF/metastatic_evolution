load data

```python
%%R

###
library(SingleCellExperiment)
library(scater)
library(SC3)
options(stringsAsFactors = FALSE)
###readin
scdata <- readRDS("scater/sc3_clustered_vis.rds")
```

visualize clusters individually

```python
%%R
####define visualization cluster
vis_cluster <- 1
####read into object
tmp <- colData(scdata)$sc3_10_clusters
tmp <- (tmp == vis_cluster)
colData(scdata)$vis_clust <- tmp
####plot
plotTSNE(
    scdata,
    colour_by = "vis_clust",
) + ggtitle(paste("cluster",toString(vis_cluster),sep=" ")) + scale_fill_manual(values=c("FALSE"="blue", "TRUE"="red"))
```
